### Word Representation (unsupervised)

   - Basic model comprises of negative log-likelihood 
       > $$ log (1 + e ^ {-s(w_{t},w_{c})}) + \sum_{n \in \mathcal{N}_{t,c}} log( 1 +  e ^ {-s(w_{t},n)})$$ 
       > where $s(w_{t},w_{c})$ maps words the context vectors to a score and is given by the dot product $s(w_{t},w_{c}) = \mathbf{u}_{w_{t}}^{T} \cdot \mathbf{v}_{w_{c}}$ , and $\mathcal{N}_{t,c}$ is a set of negative samples \
       > with 
       > $cos(\theta) = \frac{\mathbf{A} \cdot \mathbf{B}} {\lVert \mathbf{A} \rVert \cdot \lVert \mathbf{B} \rVert} $ , $\theta \propto$ objective function
   
   - Hence, the objective function is
   
        > $$ \sum_{t = 1}^{T} \left[{ \sum_{c \in \mathcal{C}_{t} } \mathit{l} (s(w_{t}, w_{c})) + \sum_{n \in \mathcal{N}_{t,c}} \mathit{l} (-s(w_{t}, n)) } \right]   $$
        > where $l : x \mapsto \log (1+e^{-x}) $
   
   - Character n-grams _ each word is represented by the sum of the vector representations of its n-grams, obtaining the resulting scoring function
       > $$ s(w,c) = \sum_{g \in \mathcal{g}_{w}} \mathbf{z}_{g}^{T} \cdot \mathbf{v}_{c} $$
       > where given a dictionary of n-grams of size $G$, $\mathcal{g}_{w} \subset \{1, ... , G\}$ denotes the set of n-grams appearing in a word $w$
       
   
    
    
### Text classification (supervised)
   - Classify texts on 

> Understanding the relationship between the predictors $X$ and the response $Y$ \
        > eg. Which predictor among $X_1, X_2, ..., X_p$ has the greatest impact on the value of the response $Y$?\
        > eg. What is the relationship between the response and each predictor?

```python
# from https://github.com/ratsgo/embedding

from soynlp.hangle import decompose, compose, character_is_korean
import re

doublespace_pattern = re.compile('\s+')

def jamo_sentence(sent):

    def transform(char):
        if char == ' ':
            return char
        cjj = decompose(char)
        if len(cjj) == 1:
            return cjj
        cjj_ = ''.join(c if c != ' ' else '-' for c in cjj)
        return cjj_

    sent_ = []
    for char in sent:
        if character_is_korean(char):
            sent_.append(transform(char))
        else:
            sent_.append(char)
    sent_ = doublespace_pattern.sub(' ', ''.join(sent_))
    return sent_


def jamo_to_word(jamo):
    jamo_list, idx = [], 0
    while idx < len(jamo):
        if not character_is_korean(jamo[idx]):
            jamo_list.append(jamo[idx])
            idx += 1
        else:
            jamo_list.append(jamo[idx:idx + 3])
            idx += 3
    word = ""
    for jamo_char in jamo_list:
        if len(jamo_char) == 1:
            word += jamo_char
        elif jamo_char[2] == "-":
            word += compose(jamo_char[0], jamo_char[1], " ")
        else:
            word += compose(jamo_char[0], jamo_char[1], jamo_char[2])
    return word
```

```python
jamo_sentence("서울특별시")
```

```python
for i in temp_model_gs.wv.most_similar(jamo_sentence("서울특벌시")):
    print(jamo_to_word(i[0]), i[1])
```
