```python
import pandas as pd
import os

import re
```

```python
root, _, files = next(iter(os.walk("./NLP_study/data")))
```

```python
for file in files:
    if file.endswith(".smi"):
        with open(root+"/"+file, "r") as f:
            lines = f.readlines()
            hangul = []
            for i in lines:
                line = i.strip()
                if not(line.startswith("<") or line.endswith("}") or line.endswith("nbsp;")):
                    question = ""
                    if "?" in line:
                        question = "?"
                    hangul.append(re.sub("[^a-z A-Z 0-9 ㄱ-ㅣ 가-힣]+", '', line)\
                                  .replace("br","").replace("BR","")\
                                  .replace("FontColorhotpink","")\
                                  .replace("FontColorpinkFontColorWHITE","")\
                                  .replace("FontColorpink","")\
                                  .replace("fontcolorhotpink","")\
                                  .replace("FontFontColor808080","")\
                                  .replace("Font","")\
                                  .replace("Color808080","")+ question
                                 )
                    
                    # "<[^<>]+>"

                    pd.Series(hangul).to_csv(file+".csv", index = False)
```

```python
root, _, files = next(iter(os.walk("./")))
```

```python
fn = 0
collated_data = pd.DataFrame(columns = ["input","label"])
#collated_data.columns = ["input","label"]

for file in files:
    if file.endswith(".csv"):
        fn += 1
        data = pd.read_csv(root+"/"+file).dropna().reset_index(drop = True)
        print(data)
        data.columns = ["input"]
        data["label"] = data["input"].values.tolist()[1:] + [""]
        collated_data = pd.concat([collated_data, data])
        print("-"*40)
        print("ADDED")
        print(data.head())
        print("COLLATED DATA")
        print(collated_data.head())
        
```
